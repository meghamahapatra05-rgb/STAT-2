1. What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to make decisions about a population using sample data. It helps us check whether an asumption is true or not.

2. What is the null hypothesis and how does this differ from alternative hypothesis?

-> Null hypothesis(Ho): Assumes no effect or no difference. Alternative hypothesis(H1): Assumes there is an effect or difference.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?

-> The significance level is the probability of rejecting a true null hypothesis. Common value: 0.05 If p-value < probability value-> reject Ho.

4. What are Type 1 and Type 2 errors? Give examples of each.

-> Type 1 error: Rejecting a true null hypothesis. Type 2 error: Not rejecting a false null hypothesis.

5. What is the difference between Z test and T test? Explain when to use each.

-> Z-test is used when population variance is known while T-test is used when population variance is unknown. Z-test has large sample size while T-test has small sample size.

8. Write a python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(50, 5, 1000)
mean = np.mean(data)
ci = stats.norm.interval(0.95, loc=mean, scale=np.std(data)/np.sqrt(len(data)))

plt.hist(data, bins=10)
plt.axvline(ci[0], color='r')
plt.axvline(ci[1], color='r')
plt.show()
```
